```python
#Helper function to preprocess the image 
def preprocess_image(image_path):
    img = image.load_img(image_path , target_size = (224,224,3))
    img  = image.img_to_array(img)
    img = np.expand_dims(img , axis =0 )
    return img
```

```python
# Creating list of unique words from the given list of words
unique_words = []
for i in words :
        unique_words.extend(i)
unique_words = list(set(unique_words))        
print(len(unique_words))
```

```python
#Vectorization 
word_to_indices = {val : i for i,val in enumerate(unique_words)}
indices_to_words = {i:val for i,val in enumerate(unique_words)}
```

```python
vocab_size = len(word_to_indices.keys())+1
print(vocab_size)
```

```python
max_len_sentence = 0
for i in sentences :
    i = i.split()
    if len(i) >  max_len_sentence:
        max_len_sentence = len(i)
print(max_len_sentence)        
        
```

```python
#Applying tokenization and padding to the given Text data
input_text  , ouput_text = [] , []
for i in range(ds.shape[0]):
    partial_seqs = []
    next_words = []
    t = ds[i][1].split()
    t = [word_to_indices[i] for i in t]
    for i in range(1, len(t)):
        partial_seqs.append(t[:i])# entire block before the current word
#         print(partial_seqs)
        next_words.append(t[i])# append the current word 
#         print(next_words)
    padded_partial_seqs = sequence.pad_sequences(partial_seqs , max_len_sentence , padding='post')
    next_words_onehot_encoded = np.zeros([len(next_words), vocab_size], dtype=np.bool)
    for i,next_word in enumerate(next_words):
        next_words_onehot_encoded[i, next_word] =1
    input_text.append(padded_partial_seqs)
    ouput_text.append(next_words_onehot_encoded)
input_text = np.asarray(input_text)
ouput_text = np.asarray(ouput_text)
print(input_text.shape)
print(ouput_text.shape)
    
    
    
```

```python
embedding_size = 128
max_len = 39
```

```python
#Building up our image model 
#model : image(2048) --> Dense
image_model =  Sequential()
image_model.add(Dense(embedding_size , input_shape=(2048,) , activation = 'relu'))
image_model.add(RepeatVector(max_len))
image_model.summary()
```

```python
#Builing up our language model 
# sequence of words --> Embeded layer -->LSTM
language_model = Sequential()
language_model.add(Embedding(input_dim = vocab_size , output_dim = embedding_size , input_length = max_len))
language_model.add(LSTM(256,return_sequences= True))
language_model.add(TimeDistributed(Dense(embedding_size)))
language_model.summary()
```

```python
#Concatenatiing the outputs of both the models and traing them togther
concatenated_model = Concatenate()([image_model.output, language_model.output])
mod = LSTM(256 , return_sequences = True)(concatenated_model)
mod = LSTM(512 , return_sequences = False)(mod)
mod = Dense(vocab_size)(mod)
out = Activation('softmax')(mod)
model = Model(inputs = [image_model.input , language_model.input] , outputs =out)
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint , EarlyStopping
model.compile(loss = 'categorical_crossentropy' , optimizer='RMSprop' , metrics = ['accuracy'])
callbacks = EarlyStopping(monitor ='val_loss' , mode ='min' , verbose =1 , patience = 10)
```

```python
history = model.fit([images , captions] ,next_words , batch_size = 512 , epochs = 200 , callbacks=callbacks)
```
